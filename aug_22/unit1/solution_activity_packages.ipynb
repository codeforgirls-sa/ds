```python
df['salary_currency'].unique()
```

```python
df['salary_currency'].unique().shape
```

```python
salary_currency_df = df.drop_duplicates(subset='salary_currency', keep='first')
salary_currency_df
```

```python
salary_currency_df = salary_currency_df[['salary_currency', 'salary', 'salary_in_usd']]
salary_currency_df.reset_index(drop=True, inplace = True)
```

```python
salary_currency_df['rate'] = [float(salary_currency_df['salary'][i])/float(salary_currency_df['salary_in_usd'][i]) 
                              for i in range (salary_currency_df.shape[0])]
```
