```python
q_1 = df["salary_in_usd"].quantile(0.25)
q_3 = df["salary_in_usd"].quantile(0.75)
iqr = q_3 - q_1

l_q = q_1 - (1.5 * iqr)
h_q = q_3 + (1.5 * iqr)
```
